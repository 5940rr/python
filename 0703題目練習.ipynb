```python
s="MCMXCIV"
roman_to_int={
    'I':1,
    'V':5,
    'X':10,
    'L':50,
    'C':100,
    'D':500,
    'M':1000
  }

n=len(s)
total=0
for i in range(n):
   value=roman_to_int[s[i]]
   if (i+1 < n and roman_to_int[s[i+1]] > value):
     total=total-value
   else:
     total=total+value
print(total)
```
