```python
x=int(input("輸入答對題數"))
score=0
if(x<=8 and x>=0):
  score=x*8
  print(score)
elif(x>8 and x<=14):
  score=64+(x-8)*6
  print(score)
else:
  print("輸入錯誤")
```
