```python
x=int(input("輸入成績"))
if(x>=90  and x<=100):
   print("級棒")
elif(x>=70 and x<=89):
   print("不錯")
elif(x>=60 and x<=69):
   print("還OK")
elif(x>=0 and x<=59):
   print("加油")
else:
   print("輸入錯誤，請重新輸入成績")
```
