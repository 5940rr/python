```python
for i in range(1,10):
  for j in range(1,10):
    print(f'{i}*{j}={i*j:1}',end=' ')
  print("")
```

```python
f=int(input("請輸入第一數字"))
s=int(input("輸入第二個數字"))
while(f>0 and s>0):
  if(f>s):
    f=f%s
  else:
    s=s%f

if(f==0):
  print("最大公因數為"+str(s))
else:
  print("最大公因數為"+str(f))
```
